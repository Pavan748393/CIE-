```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/mtcars (6).csv')
transmission_counts = df['am'].value_counts()
transmission_counts.plot(kind='bar')
plt.title('Frequency Distribution of 1')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/mtcars (6).csv')
plt.boxplot(df['mpg'])
plt.title("Boxplot of MPG")
plt.xlabel('')
plt.ylabel('MPG')
plt.show()
```
